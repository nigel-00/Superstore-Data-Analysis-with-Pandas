```python
#import the pandas library 
import pandas as pd 
#Import numpy library 
import numpy as np 
#Import matpltlib as ml
import matplotlib.pyplot as plt 
%matplotlib inline

#basic descriptive statistics on a DataFrame

df  = pd.read_csv("superstore.csv", encoding="ISO-8859-1")
descr_stats  =  df[["Sales","Quantity", "Discount","Profit"]].describe()
print(descr_stats)

#Sales of customers in California that bought a product with quantity of 14 or more in a bar chart
#This can help the company keep of customers in different locations and their amount of sales. It can 
#also help the company figure out which products have the highest quantity sold.

cal_customers= df[(df["State"]== "California") & (df["Quantity"] >= 14)]
print(cal_customers)
cust_sales = cal_customers[["Customer Name","Sales"]]
cust_sales = cust_sales.set_index("Customer Name")
cust_sales.plot(kind="bar")
plt.ylabel("Sales")


```
